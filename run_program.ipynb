```python
from get_image import get_image
from facenet2 import describe_image, show_imagebox
from matching_py import *
from Profile import *
from whispers import *
from pathlib import Path

import pickle

def run_program():
    
    with open("database.pkl", mode="rb") as database:
        loaded = pickle.load(database)
    
    profile = Profile()
    image = get_image()
#     root_dir = Path('..\image database')
    
#     i = 0
#     nodes = []
#     images = []
#     for file in root_dir.glob('*'):
#         print(file)
#         image = io.imread(file)
#         if image.shape[-1] == 4:
#             # Image is RGBA, where A is alpha -> transparency
#             # Must make image RGB.
#             image = image[..., :-1]
#         images.append(image)
        
#         descriptors, boxes, probabilities, landmarks = describe_image(image)
#         nodes.append(Node(i, neighbors=[], descriptor=descriptors[0]))
#         i += 1

#     return (nodes, images)
    descriptors, boxes, probabilities, landmarks = describe_image(image)
    
    name_list = []
    
    for descriptor in descriptors:
        name_list.append(str(matching(loaded, descriptor, 0.6)))
    
    print(name_list)
    show_imagebox(name_list, image, boxes, probabilities, landmarks)
    
    if 'Unknown :(' in name_list:
        i = name_list.index('Unknown :(')
        name = input("Please enter unknown's name")
        profile(loaded, name, descriptors[i])
    
            
    with open("database.pkl", mode="wb") as database:
        pickle.dump(loaded, database)
        
```

```python
nodes, images = run_program()
```

```python
def run_whispers(nodes):
    for test_node in nodes:
        find_neighbors(test_node, nodes)
    
    #set node labels in-place
    algorithm(nodes)
    
    labels = []
    for node in nodes:
        labels.append(node.label)
        
    return labels

labels = run_whispers(nodes)
```

```python
used = {}
with open("database.pkl", mode="rb") as database:
    loaded = pickle.load(database)
     
profile = Profile()

for i in range(len(images)):
    descriptors, boxes, probabilities, landmarks = describe_image(images[i])
    if labels[i] not in used.keys():
        show_imagebox(["Unknown :("], images[i], boxes, probabilities, landmarks)
        name = input("Please enter unknown's name >> ")
        profile(loaded, name, descriptors)
        used[labels[i]] = name
    else:
        profile(loaded, used[labels[i]], descriptors)
        
        
```
